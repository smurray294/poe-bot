```python
f = open('./terrain_img.pickle', 'rb')
terrain_img = pickle.load(f)
f.close()
plt.imshow(terrain_img)
```

```python
ret, currently_passable = cv2.threshold(cv2.convertScaleAbs(terrain_img),50,1,cv2.THRESH_BINARY)
plt.imshow(currently_passable[250:300, 425:550])
```

```python
ret, passable = cv2.threshold(cv2.convertScaleAbs(terrain_img),51,1,cv2.THRESH_BINARY)
plt.imshow(passable[250:300, 425:550]);plt.show()
plt.imshow(passable[750:1000, 750:1100]);plt.show()
plt.imshow(passable)
```

```python
kernel = np.ones((2,2), int)
eroded = cv2.erode(passable,kernel ,iterations = 2)
plt.imshow(eroded[250:300, 425:550]);plt.show()
dilated = cv2.dilate(eroded,kernel ,iterations = 1)
plt.imshow(dilated[250:300, 425:550]);plt.show()
ret, currently_passable = cv2.threshold(cv2.convertScaleAbs(dilated),0,1,cv2.THRESH_BINARY)
plt.imshow(currently_passable[250:300, 425:550]);plt.show()
```
